```python
# salary based on  the years of Experience
import pandas as pd
import numpy as np
import seaborn as sns
from warnings import filterwarnings
import statsmodels.formula.api as smf
filterwarnings('ignore')

salary_df=pd.read_csv('Salary_Data.csv')

print(salary_df.head())
print(salary_df.info())

model = smf.ols('Salary~YearsExperience',salary_df).fit()

print(model.rsquared,model.rsquared_adj)
print(model.pvalues,model.tvalues)

sns.regplot(salary_df.YearsExperience,salary_df.Salary)

```

```python
#Delivery_time -> Predict delivery time using sorting time
import statsmodels.formula.api as sfa

delivery_df=pd.read_csv('delivery_time.csv')
print(delivery_df.head())
print(delivery_df.info())
x=delivery_df.iloc[:,0]
y=delivery_df.iloc[:,1]

model1=sfa.ols("y~x",data=delivery_df).fit()

print(model1.pvalues,model1.tvalues)
print(model1.rsquared_adj,model1.rsquared)

sns.regplot(x,y)
```
